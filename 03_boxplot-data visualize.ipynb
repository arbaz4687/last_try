```python
import seaborn as sns 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
kashti=sns.load_dataset("titanic")
kashti
sns.boxplot(x="survived"
      ,y="age",
      data=kashti)
plt.show()
```

```python
# show means
sns.boxplot(x="survived"
      ,y="age",showmeans=True,
      data=kashti)
plt.show()
```

```python
sns.boxplot(x="survived",
      y="age",showmeans=True,
            meanprops={"marker":"+","markersize":"12",
                       "markeredgecolor":"red"},
      data=kashti)
plt.xlabel("how many survived",size=20,weight="bold"),
plt.ylabel("what is age",size=20,weight="bold"),
plt.title("mott ka hisab",size=20,weight="bold"),
plt.show()
```
