```python
import cv2 as cv
import numpy as np
img=cv.imread("resorces/download.jpg")
# convert into hsv(Hue,saturation,value)
hsv_img=cv.cvtColor(img,cv.COLOR_BGR2HSV)
# create a function OF SILDER:
def slider():
     pass
path="resorces/download.jpg"# USE only add new through path
cv.namedWindow("Bars") # use or new window (image show)
cv.resizeWindow("Bars",640,300)

# create a trackbar on bar window that is created above 
cv.createTrackbar("Hue minimum","Bars",0,179,slider)# maximum value of hue(180)
cv.createTrackbar("Hue maximum","Bars",179,179,slider)
cv.createTrackbar("sat minimum","Bars",0,255,slider)# maximum value of satur(255)
cv.createTrackbar("sat maximum","Bars",255,255,slider)
cv.createTrackbar("val minimum","Bars",0,255,slider)
cv.createTrackbar("val maximum","Bars",255,255,slider)
## when we change the trackbar so a value is not increase and decrease in python
while(True):
    img=cv.imread(path)
    hsv_img=cv.cvtColor(img,cv.COLOR_BGR2HSV)
    hue_min=cv.getTrackbarPos("Hue minimum","Bars")
    hue_max=cv.getTrackbarPos("Hue maximum","Bars")
    sat_min=cv.getTrackbarPos("sat minimum","Bars")
    sat_max=cv.getTrackbarPos("sat maximum","Bars")
    val_min=cv.getTrackbarPos("val minimum","Bars")
    val_max=cv.getTrackbarPos("val maximum","Bars")
    print(hue_min, hue_max,sat_min,sat_max,val_min,val_max)
    # apply this function on image 
    # creat an ipper and lower arrays 

    lower=np.array([hue_min,sat_min,val_min])
    upper=np.array([hue_max,sat_max,val_max])
    # create an mas of image 
    mask_img=cv.inRange(hsv_img,lower,upper)
    # final output show on image 
    out_img=cv.bitwise_and(img,img,mask=mask_img)
    
    cv.imshow("org",img)
    cv.imshow("hsv",hsv_img)
    cv.imshow("mask",mask_img)
    cv.imshow("out",out_img)
    if cv.waitKey(1) & 0xff ==ord("q"):
        break
    
cv.destroyAllWindows()
```
