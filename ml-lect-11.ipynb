```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import online data
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
#input variables ya feature(X)
digits.data.shape
X=digits.data

## 1797 is an images and 64 image size =8*8
```

```python
### output labels(y)
digits.target.shape
y=digits.target
```

```python
plt.figure(figsize=(20,4))
for index,(image,label) in enumerate(zip(digits.data[0:20],digits.target[0:20])): ## double 0:20 use for input paramets and output parameters and for loop used
   plt.subplot(1,20,index+1)
   plt.imshow(np.reshape(image,(8,8)),cmap=plt.cm.gray)
   plt.title(label,fontsize=20)
```

#### command help(plt) define how to draw a plot

```python
## spliting data 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=True)
```

```python
X_train.shape
```

```python
print("train input data:",X_train.shape)
print("train out data:",X_test.shape)
print("train input data:",y_train.shape)
print("train out data:",X_test.shape)
```

```python
## model creation 
from sklearn.linear_model import LogisticRegression
model=LogisticRegression().fit(X_train,y_train)
model
```

```python
model.predict(X_test)
```

```python
model.predict(X_test[0:10])
prediction=model.predict(X_test)
prediction
```

```python
### accuracy checking
score=model.score(X_test,y_test)
score
```

```python
## confudsion matrix
from sklearn import metrics
cm=metrics.confusion_matrix(y_test,prediction)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(9,9))
sns.heatmap(cm,annot=True,fmt=".3f",linewidths=.5, square=True,cmap='Spectral');
plt.ylabel("Actual output");
plt.xlabel("predicted out");
all_sample_title='Accuracy score:{0}'.format(score)
plt.title(all_sample_title,size=15);
```
