```python
### R2 is use for checking the accuracy of model 
import numpy 
from sklearn.metrics import r2_score
x=[89,43,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y=[21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47]
mymodel =numpy.poly1d(numpy.polyfit(x,y,3))
print(r2_score(y,mymodel(x)))
```

```python
 # import onine data_set
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("salary.csv")
df.head()
```

```python
df["gender"]=df["gender"].replace("Male",1)
df["gender"]=df["gender"].replace("female",0)
```

```python
X=df[["Age"]]
y=df["Salary"]
```

```python
X.head()
```

```python
y.head()
```

```python
### spliting of data 80 by 20 role
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=True)
```

```python
## model fitting linear regression line to data set
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,y)
## visulization of  linear regression result
def viz_linear(): ## viz_linear function
    plt.scatter(X,y,color="red")
    plt.plot(X,lin_reg.predict(X),color="blue")
    plt.title("truth or Bluff")
    plt.xlabel("position")
    plt.ylabel("salary")
    plt.show()
    return
viz_linear()
```

```python
## fiting polynomial regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
X_poly=poly_reg.fit_transform(X) ### convertion of X dataset  into polynomial
pol_reg=LinearRegression()
pol_reg.fit(X_poly,y)

## vizualize the polynomial plot
def viz_polynomial():
    plt.scatter(X,y,color="red")
    plt.plot(X,pol_reg.predict(poly_reg.fit_transform(X)),color="blue")
    plt.title("true polynomial figure ")
    plt.xlabel("position level")
    plt.ylabel("salary ")
    plt.show()
    return
viz_polynomial()
```

```python
### prediction the result using linear regression
predict1=lin_reg.predict([[21.0]])
predict1
```

```python
predict2=pol_reg.predict(poly_reg.fit_transform([[21.0]]))
predict2
```

```python
print('LinearRegression=',predict1)
print('polylinearRegression=',predict2)
print("difference =",predict1-predict2)
```
