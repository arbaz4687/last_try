### coorelation
 - parson's correlation coefficient
    -  each sample normaly distributed 
    -  each sample have the same variance
- H0:two samples are independent(if h0 selected independent )
- H1:there are dependency between samples (h1 selected dependent )

```python
from scipy.stats import pearsonr
data1=[0.873,2.827,0.121,-0.954,-1.436,0.360,-1.478,-1.637,-1.869]
data2=[0.353,3.517,0.125,-7.545,0.555,-1.536,3.350,-1.578,-3.537]
stat,p=pearsonr(data1,data2)
print("state_test data=",stat)
print("p_value test data=",p)
if p > 0.05:
 print("probability independent")
else:
 print("probabilitydependent(corerelation exist)")
```

### data using spearmanr
- normality does not matter(test is perform on continous data)

```python
### find correlationship between data using spearmanr
from scipy.stats import spearmanr
data1=[0.873,2.827,0.121,-0.954,-1.436,0.360,-1.478,-1.637,-1.869]
data2=[0.353,3.517,0.125,-7.545,0.555,-1.536,3.350,-1.578,-3.537]
stat,p=spearmanr(data1,data2)
print("state_test data=",stat)
print("p_value test data=",p)
if p > 0.05:
 print("probability independent")
else:
 print("probabilitydependent(corerelation exist)")
```

## chi-square method(kayi test)
 - in chi - square we are performed operation on categorical data (while in correlaton ship test perform on continues data ) only rank data are used for testing(table data get after plus/minus)
    - H0:two samples are independent(if h0 selected independent )
    - H1:there are dependency between samples (h1 selected dependent )

```python
from scipy.stats import chi2_contingency
table=[[10,20,30],[6,9,17]]
stat,p,dof,expected=chi2_contingency(table)
print("state_test data=",stat)
print("p_value test data=",p)
if p > 0.05:
 print("probability independent")
else:
 print("probabilitydependent(corerelation exist)")
```

### parametric statistics hypothisitesstudent-T test(unpaired )
 - used two find the relationship between two different data sets
  

```python
from scipy.stats import ttest_ind
#data1=[0.873,2.827,0.121,-0.954,-1.436,0.360,-1.478,-1.637,-1.869]
#data2=[0.353,3.517,0.125,-7.545,0.555,-1.536,3.350,-1.578,-3.537]
data1=ks["age"].dropna()
data2=ks["fare"].dropna()
stat,p=ttest_ind(data1,data2)
print("state_test data=",stat)
print("p_value test data=",p)
if p > 0.05:
 print("probability same data ")
else:
 print("probability different data ")
```

### Pair student  t_test
- if we get the two things in same sample.means in wich data lenght is not matter 
- in which we compare the data set with same sample.

```python
from scipy.stats import ttest_rel
data1=[0.873,2.827,0.121,-0.954,-1.436,0.360,-1.478,-1.637,-1.869]
data2=[0.353,3.517,0.125,-7.545,0.555,-1.536,3.350,-1.578,-3.537]
stat,p=ttest_rel(data1,data2)
print("state_test data=",stat)
print("p_value test data=",p)
if p > 0.05:
 print("probability same data ")
else:
 print("probability different data ")
```

### Anova (one way)
- are used for get the result in more than two dataset

```python
from scipy.stats import f_oneway
data1=[0.873,2.827,0.121,-0.954,-1.436,0.360,-1.478,-1.637,-1.869]
data2=[0.353,3.517,0.125,-7.545,0.555,-1.536,3.350,-1.578,-3.537]
data3=[0.873,2.827,0.121,-0.954,-1.436,0.360,-1.436,0.360,-1.478,-1.637,-1.869]
stat,p=f_oneway(data1,data2,data3)
print("state_test data=",stat)
print("p_value test data=",p)
if p > 0.05:
 print("probability same data ")
else:
 print("probability different data ")
```
